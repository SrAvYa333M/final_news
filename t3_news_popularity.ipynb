```python
#drop null value columns
df.dropna(inplace=True)
df
```

```python
df.isnull().sum()
```

```python
for col in ['PublishDate', 'Headline', 'SentimentTitle','SentimentHeadline']:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Calculate the correlation matrix, excluding non-numeric columns
correlation_matrix = df.corr(numeric_only=True)
correlation_matrix

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(df[['Facebook']], df['GooglePlus'])
```

```python
plt.scatter(df['Facebook'], df['GooglePlus'])
plt.xlabel('Facebook')
plt.ylabel('Googleplus')
plt.title('Scatter Plot')
plt.show()
```

```python

plt.plot(df['Facebook'], model.predict(df[['Facebook']]), color='red')
```

```python
#draw best fit line
plt.scatter(df['Facebook'], df['GooglePlus'])
plt.xlabel('Facebook')
plt.ylabel('Googleplus')
plt.title('Scatter Plot')
plt.plot(df['Facebook'], model.predict(df[['Facebook']]), color='red')
```

```python
#find mean square error
from sklearn.metrics import mean_squared_error
mean_squared_error(df['GooglePlus'], model.predict(df[['Facebook']]))
```
